```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# cmap => colormap
plt.rcParams["figure.figsize"] = 10, 5

r_x = np.random.randn (10000)
r_y = np.random.randn (10000)

fig, ax = plt.subplots (1,2)
fig.text (.35, 0, 'Gaussian distributed random numbers.')

ax[0].hist2d (r_x, r_y)
ax[1].hist (r_x, bins='auto', histtype='step')

plt.show ()
```

```python
w, h = 256, 256 
noise = np.random.randn (w, h, 3)

plt.imshow (noise, cmap='gray')
plt.show ()
```
