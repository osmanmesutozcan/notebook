```python
import cv2
import scipy.ndimage
import numpy as np
import matplotlib.pyplot as plt
```

```python
r_x = np.random.randn (10000)
r_y = np.random.randn (10000)

fig, ax = plt.subplots (1,2, figsize=(20, 10))
fig.text (.35, 0, 'Gaussian distributed random numbers.')

ax[0].hist2d (r_x, r_y)
ax[1].hist (r_x, bins='auto', histtype='step')

plt.show ()
```

```python
w, h = 256, 256 

noise = np.random.randn (w, h)
orig_im = scipy.ndimage.imread('data/cat.jpg', 'L')
im = cv2.resize(orig_im, (w, h), interpolation=cv2.INTER_CUBIC)

def blend_image (im1, im2, scale):
    return (im1 / 1 - scale) + (im2 / scale)

fig, ax = plt.subplots(1,3, figsize=(20,10))

ax[0].imshow (noise, cmap='gray')
ax[1].imshow (im, cmap='gray')
ax[2].imshow (blend_image(im, noise, 0.08), cmap='gray')

plt.show ()
```
