```python
import cv2
import matplotlib.pyplot as plt

from computer_vision.color_transfer import color_transfer


source = cv2.imread("./data/color_transfer_source.jpg")
target = cv2.imread("./data/color_transfer_target.jpg")
transfer = color_transfer(source, target)

fig, axis = plt.subplots(3, 1, figsize=(8, 8))
ax = axis.ravel()

ax[0].imshow(cv2.cvtColor(source.astype('uint8'), cv2.COLOR_BGR2RGB))
ax[0].set_title('Source')

ax[1].imshow(cv2.cvtColor(target.astype('uint8'), cv2.COLOR_BGR2RGB))
ax[1].set_title('Target')

ax[2].imshow(transfer)
ax[2].set_title('Transferred')
```
